```python
data_yark_sk3 = pd.read_csv('[pro] bennu clone_0_sx0.00_sy0.00_sz1.00.csv')
df_yark_sk3 = pd.DataFrame(data_yark_sk3)

data_yark_sk5 = pd.read_csv('[pro] bennu clone_0_sx-1.00_sy0.00_sz0.00.csv')
df_yark_sk5 = pd.DataFrame(data_yark_sk5)
```

```python
# Plot a non-yarkovsky vs yarkovsky prograde (skema 1 dan 2)

x_ae = df_yark_sk3['Time (yr since J2000)']
y3a = df_yark_sk3['a (AU)']
y3e = df_yark_sk3['e']

y5a = df_yark_sk5['a (AU)']
y5e = df_yark_sk5['e']

fig, (ax1a, ax2a) = plt.subplots(2, 1, figsize=(10, 10))  # 1 row, 2 plots

# --- Top plot ---
line3a, = ax1a.plot(x_ae, y3a, 'k-', label='Sumbu semi-mayor')
ax1a.set_ylabel('Sumbu semi-mayor (SA)')  # remove if you want shared legend only
#ax1.tick_params(axis='y', labelcolor='k')

ax1e = ax1a.twinx()
line3e, = ax1e.plot(x_ae, y3e, 'c-', label='Eksentrisitas')
ax1e.set_ylabel('Eksentrisitas')
# ax1b.tick_params(axis='y', labelcolor='c')

# Combine both lines into one legend
ax1a.legend(handles=[line3a, line3e], loc='upper left')
#ax1a.set_ylim(0.4, 1.5)
#ax1e.set_ylim(0.4, 1.5)
ax1a.grid(True)
ax1a.set_title("Sumbu semi-mayor dan eksentrisitas dengan efek Yarkovsky pada sumbu putar (0, 0, 1)")

# --- Bottom plot ---
line5a, = ax2a.plot(x_ae, y5a, 'k-', label='Sumbu semi-mayor')
ax2a.set_ylabel('Sumbu semi-mayor (SA)')
#ax2a.tick_params(axis='y', labelcolor='k')

ax2e = ax2a.twinx()
line5e, = ax2e.plot(x_ae, y5e, 'm-', label='Eksentrisitas')
ax2e.set_ylabel('Eksentrisitas')
# ax2b.tick_params(axis='y', labelcolor='m')

ax2a.legend(handles=[line5a, line5e], loc='upper left')
#ax2a.set_ylim(0.4, 1.5)
#ax2e.set_ylim(0.4, 1.5)
ax2a.grid(True)
ax2a.set_title("Sumbu semi-mayor dan eksentrisitas dengan efek Yarkovsky pada sumbu putar (-1, 0, 0)")
ax2a.set_xlabel("Waktu (dalam tahun)")

plt.tight_layout()
plt.savefig("ae skema 3 5.png", dpi=300)
plt.show()
```
